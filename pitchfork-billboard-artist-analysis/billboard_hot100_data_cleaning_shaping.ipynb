# Data Reshaping

BB - total weeks on chart, by artist (group by: artist + track, max: weeks on board > group by: artist, sum: weeks on board)

```python
bb_weeks_track = bb.pivot_table(index="song", values=["artist", "weeks-on-board"], aggfunc="max")

bb_weeks = bb_weeks_track.pivot_table(index="artist", values="weeks-on-board", aggfunc="sum")
```

```python
bb_weeks.sort_values(by="weeks-on-board", ascending=False, inplace=True)
```

```python
# DONE = bb_weeks
```

BB - average top ranking, by artist (group by: artist + track, max: peak rank > group by: artist, avg: peak rank #1

```python
bb_peak_rank_track = bb.pivot_table(index="song", values=["artist", "peak-rank"], aggfunc="min")

bb_peak_rank = bb_peak_rank_track.pivot_table(index="artist", values="peak-rank", aggfunc="mean")
```

```python
bb_peak_rank.sort_values(by="peak-rank", ascending=True, inplace=True)
```

```python
bb_peak_rank.head(500)
```

```python
# DONE = bb_peak_rank
```

BB - average top ranking, by artist (group by: artist + track, avg: rank > group by: artist, avg: rank) #2

```python
bb_rank_track = bb.pivot_table(index=["song", "artist"], values="rank", aggfunc="mean")

bb_rank_track.reset_index(inplace=True)
```

```python
bb_rank = bb_rank_track.pivot_table(index="artist", values="rank", aggfunc="mean")

bb_rank.sort_values(by="rank", ascending=True, inplace=True)
```

```python
bb_rank.head(500)
```

```python
# DONE = bb_rank
```

# Data Merging

```python
bb_all = bb_weeks.merge(bb_peak_rank, how="inner", on="artist").merge(bb_rank, how="inner", on="artist")
```

```python
bb_all.reset_index(inplace=True)
```
