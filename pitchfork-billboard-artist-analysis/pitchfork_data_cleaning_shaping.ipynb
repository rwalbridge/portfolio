# PF(all) Data Reshaping

PF - average review, by artist (group by: artist, avg: rating)

```python
pf_ka_ws
```

```python
pf_ka_ws_scores = pf_ka_ws.pivot_table(index="artist", values="score", aggfunc="mean")
```

```python
pf_ka_ws_scores.sort_values(by="score", ascending=False, inplace=True)
```

```python
pf_ka_ws_scores
```

```python
# DONE = pf_ka_scores
```

PF - # of albums, by artist (group by: artist, count: album name)

```python
pf_ka_ws_no_albums = pf_ka_ws.pivot_table(index="artist", values="album", aggfunc="count")
```

```python
pf_ka_ws_no_albums.sort_values(by="album", ascending=False, inplace=True)
```

```python
pf_ka_ws_no_albums
```

```python
# DONE = pf_ka_no_albums
```

# PF(ka) Data Merging

```python
pf_all = pf_ka_ws_scores.merge(pf_ka_ws_no_albums, how="inner", on="artist")
```

```python
pf_all.sort_values(by="album", ascending=False, inplace=True)
```

```python
pf_all.reset_index(inplace=True)
```
